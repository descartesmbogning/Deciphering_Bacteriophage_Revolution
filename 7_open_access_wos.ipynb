## Open access expand

```python
open_access_df_expand = open_access_df['Open Access Designations'].str.split(', ', expand=True)
open_access_df_expand
```

## Get dummies on open access

```python
# get dummies to have contries on columns
from functools import reduce
dfs = [pd.get_dummies(open_access_df_expand[col]) for col in open_access_df_expand]
d = reduce(lambda x,y : x.add(y, fill_value=0), dfs)

# data cleaning
cols = [c for c in d.columns if c[-1:] != '.']
df=d[cols]
cols = [c for c in df.columns.to_list() if len(c) != 1]
df=df[cols]

# Save data
df.to_pickle(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/open_access_wos_expand.pkl')
df.to_csv(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/open_access_wos_expand.csv')
#
df
```

## Open access freq

```python
# Create an empty list
Open_access_list =[]

my_list = open_access_df_expand.columns.to_list
# Iterate over each row
for index, rows in open_access_df_expand.iterrows():
    # Create list for the current row
    for i in [range(len(open_access_df_expand.columns))]:
        my_list=rows[i]
        for el in rows[i]:
            if el != None:
                Open_access_list.append(el)


# result
Open_access_list = pd.DataFrame(Open_access_list)
Open_access_freq = Open_access_list.value_counts().to_frame()
Open_access_freq = Open_access_freq.rename(columns={0: "Number per citation"})
Open_access_freq = Open_access_freq.reset_index()
Open_access_freq = Open_access_freq.rename(columns={0: "Open Access Designations"})
Open_access_freq['Proportion (%)'] = Open_access_freq['Number per citation']/open_access_df_expand.shape[0]*100

# Save data
Open_access_freq.to_pickle(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/table/open_access_wos_freq.pkl')
Open_access_freq.to_csv(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/table/open_access_wos_freq.csv')

Open_access_freq#.shape
```
